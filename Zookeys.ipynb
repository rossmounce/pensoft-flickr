#AUTHORS YEAR TITLE JOURNAL URL

```python
ayt = soup.find_all('div', class_="authors")
print (ayt)
```

```python
file = open("./"+chomped+"/bibinfo.txt", "wb")
```

```python
from string import digits
tit = soup.find('title')
title = tit.string.strip().encode("utf8")
doibit = soup.find('div', class_="doiNumber")
outit = (doibit.string).strip().replace("doi: ","http://dx.doi.org/").encode("utf8")
blah = unicode(outit.strip().replace("doi: ","http://dx.doi.org/"),'utf-8')
#print (title+". "+outit)
#file.write(title+". "+outit)
for bl in ayt:
        authors = (bl.get_text()).encode("utf8")
        cheese = (authors.translate(None, digits)).replace("†","").replace("‡","").replace("§","").replace("|","").replace("¶","").replace("#","").replace(",,",",")
        #print (re.sub('[\W_]+', ' ', authors)) #Carmelo And jar1 Carles Hernando2 Ignacio Ribera3 
        print (unicode(cheese,'utf-8')+" "+title+". "+blah)
        #file.write("Source: "+cheese+" "+title+". "+outit)
```

```python
file.flush()
file.close()
```
